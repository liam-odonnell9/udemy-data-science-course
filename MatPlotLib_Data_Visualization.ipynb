```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
## FUNCTIONAL METHOD
plt.plot(x,y,)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])


axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X^2')

ax.plot(x,x**3, label = 'X^3')

ax.legend(loc=(0.1,0.1))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,y,color='purple',lw=3,alpha=1,ls='-',marker='o',markersize=20,
#        markerfacecolor = 'blue',markeredgewidth=3,markeredgecolor='g')

ax.plot(x,y,color='purple',lw=3,ls='--')
ax.set_xlim([0,2])
ax.set_ylim([0,2])
```
